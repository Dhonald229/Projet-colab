**Projet D'étude d'investissement**

```python
import pandas as pd
```

```python
import yfinance as yf
```

```python
prix=yf.Ticker("AAPL").history(period="5y")
prix
```

```python
prix_triés=prix.drop_duplicates().dropna()
prix_triés
```

```python
prix_triés.dtypes
```

```python
incohérences=prix_triés[prix_triés["Low"]>prix_triés["High"]]
incohérences
```

```python
prix_triés_vn=prix_triés[prix_triés["Volume"]==0]
prix_triés_vn
```

Calcul du rendement journalier:

```python
prix_triés["rendement journalier"]=prix_triés["Close"].pct_change()*100
prix_triés
```

```python
pt=prix_triés
rdm_max=round(pt['rendement journalier'].max(), 2)
volume_correspondant=pt[pt['rendement journalier']==pt['rendement journalier'].max()]
print("le rendement journalier maximal est",rdm_max,"%")
```

```python
ptdj=pt.tail(90)
ptdj['rendement journalier'].mean()*100
```
